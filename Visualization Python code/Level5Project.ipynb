```python
import matplotlib.pyplot as plt
from adjustText import adjust_text

plt.figure(figsize=(8, 6))
plt.scatter(df1['Research'], df1['% of GDP'])

texts = []
for i, country in enumerate(df1['Country']):
    texts.append(
        plt.text(
            df1['Research'].iloc[i],
            df1['% of GDP'].iloc[i],
            country,
            fontsize=8
        )
    )


adjust_text(
    texts,
    arrowprops=dict(arrowstyle='-', color='gray', lw=0.5)
)

plt.xlabel('Research Score')
plt.ylabel('R&D Spending (% of GDP)')
plt.title('Relationship between Research Strength and R&D Spending')

plt.savefig(
    r"C:\Personal\Susi\Barcelona_Activa\Project\OutputImage\ai_index_Research",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# Example mapping of hemisphere for countries
northern_countries = [
    'United States','China','Germany','Sweden','Israel','South Korea', 
    'Japan','United Kingdom','Canada','France','India'  # etc.
]

df['Hemisphere'] = df['Country'].apply(
    lambda c: 'Northern' if c in northern_countries else 'Southern'
)

# Count how many are in each hemisphere
counts = df['Hemisphere'].value_counts()
print(counts)

# Bar chart
counts.plot(kind='bar')
plt.title('Top AI Countries by Hemisphere')
plt.xlabel('Hemisphere')
plt.ylabel('Number of Countries')
plt.show()
```
