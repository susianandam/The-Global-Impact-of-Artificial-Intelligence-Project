```python
1.	Which countries score highest in Talent, Infrastructure, and Research?
```

```python
import numpy as np
import matplotlib.pyplot as plt

top_countries = df1.sort_values("Total score", ascending=False).head(10)

metrics = ["Talent", "Infrastructure", "Research"]
x = np.arange(len(top_countries))
width = 0.25

plt.figure(figsize=(12, 6))

for i, metric in enumerate(metrics):
    plt.bar(x + i*width, top_countries[metric], width, label=metric)

plt.xticks(x + width, top_countries["Country"], rotation=45, ha="right")
plt.ylabel("Score")
plt.xlabel("Country")
plt.title("Talent, Infrastructure, and Research Scores (Top 10 Countries)")
plt.ylim(0, 105)
plt.legend()
plt.tight_layout()

plt.savefig(
    r"C:\Personal\Susi\Barcelona_Activa\Project\OutputImage\ai_index_grouped_bar",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
Q1 = df1["Total score"].quantile(0.25)
Q3 = df1["Total score"].quantile(0.75)
IQR = Q3 - Q1

outliers = df1[
    (df1["Total score"] < Q1 - 1.5 * IQR) |
    (df1["Total score"] > Q3 + 1.5 * IQR)
]

print(outliers[["Country", "2024", "Total score"]])
```

```python
plt.figure(figsize=(8, 6))

# Normal points
plt.scatter(
    df1["2024"],
    df1["Total score"],
    alpha=0.6,
    label="Countries"
)

# Outliers
plt.scatter(
    outliers["2024"],
    outliers["Total score"],
    edgecolors="black",
    s=120,
    label="Outliers"
)

# Label outlier countries
for _, row in outliers.iterrows():
    plt.text(
        row["2024"],
        row["Total score"],
        row["Country"],
        fontsize=9,
        ha="left",
        va="bottom"
    )

plt.xlabel("GDP (2024)")
plt.ylabel("AI Index Total Score")
plt.title("GDP vs AI Index Total Score (IQR Outliers)")
plt.legend()
plt.tight_layout()
plt.show()
```
