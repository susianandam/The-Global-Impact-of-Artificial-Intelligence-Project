```python
import pandas as pd
df2 = pd.read_csv("C:/Personal/Susi/Barcelona_Activa/Project/Dataset/GDP_country.csv",skiprows=4)
```

```python
df3= df2.copy()
```

```python
df3.info()
```

```python
country_mapping = {
    'United States':'USA',
    'France':'France',
    'Australia':'Australia',
      'Brazil'  :'Brazil',
       'China'  :'China',
      'Canada' :'Canada',
      'Korea, Rep.'  :'South Korea',
      'India'   :'India',
       'Germany' :'Germany', "United Kingdom":"UK",
}
```

```python
df3['Country Name']=df3['Country Name'].replace(country_mapping)
```

```python
df3['Country Name'].unique()
```

```python
selected_df3 = df3[['Country Name','Indicator Name','2023','2024']]
```

```python
selected_df3
```

```python
gdp_long = selected_df3.melt(
    id_vars='Country Name',
    value_vars=['2023', '2024'],
    var_name='year',
    value_name='GDP'
)
```

```python
gdp_long['year'] = gdp_long['year'].astype(int)
```

```python
gdp_long.rename(columns={'Country Name': 'country'}, inplace=True)
```

```python
gdp_long
```

```python
gdp_long.isnull().sum().sum()
```

```python
gdp_long[gdp_long['GDP'].isnull()]
```

```python
gdp_long[gdp_long['GDP'].isnull()][['country', 'year']].head()
```

```python
gdp_long = gdp_long.dropna(subset=['GDP'])
```

```python
gdp_long.isnull().sum().sum()
```

```python
merged_df1 = pd.merge(
    gdp_long,
    df1,
    on=['country', 'year'],
    how='inner'
)
```
